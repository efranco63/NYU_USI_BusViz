```python
%matplotlib inline
from pandas import DataFrame as df
```

api pings

```python
speed=df.from_csv('MTA-Bus-Time_.2014-08-04.txt',sep='\t',index_col=None)
```

```python
B46_allpings=speed[speed['inferred_route_id']=='MTA NYCT_B46'].sort(columns=['vehicle_id','time_received'],ascending=True)
```

```python
B46_allpings['inferred_trip_id']=B46_allpings['inferred_trip_id'].apply(lambda x:x.split('_')[2].split('-')[1]+'-'+x.split('_')[2].split('-')[-1]+'_'+'_'.join(x.split('_')[-2:]))
```

trips-shape

```python
trips=df.from_csv('gtfs-0804/trips.txt')
trips['trip_id']=trips['trip_id'].apply(lambda x: x.split('-')[1]+'-'+x.split('-')[-1])
```
